```python
"""
üèÜ ATIVIDADE PR√ÅTICA: FUNDAMENTOS DE PYTHON COM COPILOT
Desenvolvido para o Google Colab
"""

# ==================== EXERC√çCIO 1: CONCATENANDO DADOS ====================
print("=" * 50)
print("EXERC√çCIO 1 - CONCATENANDO DADOS üêæ")
print("=" * 50)

def concatenar_dados():
    """Recebe dois dados e os concatena em uma √∫nica string."""
    print("\nüìù Vamos concatenar duas informa√ß√µes!")
    dado1 = input("Digite a primeira informa√ß√£o: ")
    dado2 = input("Digite a segunda informa√ß√£o: ")

    resultado = dado1 + dado2
    print(f"\n‚úÖ Resultado da concatena√ß√£o: {resultado}")
    return resultado

# Executando o exerc√≠cio 1
resultado_concatenacao = concatenar_dados()

# ==================== EXERC√çCIO 2: REPETINDO TEXTOS ====================
print("\n" + "=" * 50)
print("EXERC√çCIO 2 - REPETINDO TEXTOS ‚úèÔ∏è")
print("=" * 50)

def repetir_texto():
    """Repete uma string um n√∫mero determinado de vezes."""
    print("\nüîÑ Vamos repetir um texto v√°rias vezes!")
    texto = input("Digite o texto a ser repetido: ")

    while True:
        try:
            vezes = int(input("Digite quantas vezes repetir (n√∫mero inteiro): "))
            if vezes >= 0:
                break
            else:
                print("‚ö†Ô∏è Por favor, digite um n√∫mero n√£o negativo.")
        except ValueError:
            print("‚ö†Ô∏è Entrada inv√°lida! Digite um n√∫mero inteiro.")

    resultado = texto * vezes
    print(f"\n‚úÖ Texto repetido {vezes} vezes: {resultado}")
    return resultado

# Executando o exerc√≠cio 2
resultado_repeticao = repetir_texto()

# ==================== EXERC√çCIO 3: OPERA√á√ïES MATEM√ÅTICAS ====================
print("\n" + "=" * 50)
print("EXERC√çCIO 3 - OPERA√á√ïES MATEM√ÅTICAS SIMPLES üìê")
print("=" * 50)

def operacoes_matematicas():
    """Realiza opera√ß√µes matem√°ticas b√°sicas com dois n√∫meros."""
    print("\nüßÆ Vamos realizar opera√ß√µes matem√°ticas!")

    while True:
        try:
            num1 = float(input("Digite o primeiro n√∫mero: "))
            num2 = float(input("Digite o segundo n√∫mero: "))
            break
        except ValueError:
            print("‚ö†Ô∏è Entrada inv√°lida! Digite n√∫meros v√°lidos.")

    print("\nüìä Resultados das opera√ß√µes:")
    print(f"   Soma: {num1} + {num2} = {num1 + num2}")
    print(f"   Subtra√ß√£o: {num1} - {num2} = {num1 - num2}")
    print(f"   Multiplica√ß√£o: {num1} √ó {num2} = {num1 * num2}")

    if num2 != 0:
        print(f"   Divis√£o: {num1} √∑ {num2} = {num1 / num2:.2f}")
    else:
        print("   Divis√£o: N√£o √© poss√≠vel dividir por zero!")

    return num1 + num2, num1 - num2, num1 * num2, num1 / num2 if num2 != 0 else None

# Executando o exerc√≠cio 3
resultados_operacoes = operacoes_matematicas()

# ==================== EXERC√çCIO 4: VERIFICANDO PAR OU √çMPAR ====================
print("\n" + "=" * 50)
print("EXERC√çCIO 4 - VERIFICANDO N√öMEROS PARES E √çMPARES üßÆ")
print("=" * 50)

def verificar_par_impar():
    """Verifica se um n√∫mero √© par ou √≠mpar."""
    print("\nüî¢ Vamos verificar se um n√∫mero √© par ou √≠mpar!")

    while True:
        try:
            numero = int(input("Digite um n√∫mero inteiro: "))
            break
        except ValueError:
            print("‚ö†Ô∏è Entrada inv√°lida! Digite um n√∫mero inteiro.")

    # Verifica√ß√£o usando operador de m√≥dulo (%)
    if numero % 2 == 0:
        resultado = "PAR"
        print(f"\n‚úÖ O n√∫mero {numero} √© {resultado}!")
    else:
        resultado = "√çMPAR"
        print(f"\n‚úÖ O n√∫mero {numero} √© {resultado}!")

    return resultado

# Executando o exerc√≠cio 4
resultado_par_impar = verificar_par_impar()

# ==================== EXERC√çCIO 5: CALCULANDO M√âDIA ====================
print("\n" + "=" * 50)
print("EXERC√çCIO 5 - CALCULANDO M√âDIA DE NOTAS üìö")
print("=" * 50)

def calcular_media():
    """Calcula a m√©dia de tr√™s notas."""
    print("\nüìà Vamos calcular a m√©dia de tr√™s notas!")

    notas = []
    for i in range(1, 4):
        while True:
            try:
                nota = float(input(f"Digite a {i}¬™ nota (0-10): "))
                if 0 <= nota <= 10:
                    notas.append(nota)
                    break
                else:
                    print("‚ö†Ô∏è Nota deve estar entre 0 e 10.")
            except ValueError:
                print("‚ö†Ô∏è Entrada inv√°lida! Digite um n√∫mero.")

    media = sum(notas) / len(notas)

    print(f"\nüìä Notas informadas: {notas}")
    print(f"‚úÖ M√©dia calculada: {media:.2f}")

    if media >= 6:
        print("üéâ Situa√ß√£o: APROVADO!")
    elif media >= 4:
        print("‚ö†Ô∏è Situa√ß√£o: RECUPERA√á√ÉO!")
    else:
        print("‚ùå Situa√ß√£o: REPROVADO!")

    return media

# Executando o exerc√≠cio 5
media_calculada = calcular_media()

# ==================== EXERC√çCIO 6: VERIFICANDO PAL√çNDROMOS ====================
print("\n" + "=" * 50)
print("EXERC√çCIO 6 - VERIFICANDO PAL√çNDROMOS üîÑ")
print("=" * 50)

def verificar_palindromo():
    """Verifica se uma palavra √© um pal√≠ndromo."""
    print("\nüîÑ Vamos verificar se uma palavra √© um pal√≠ndromo!")
    print("üí° Pal√≠ndromo: palavra que se l√™ da mesma forma de tr√°s para frente.")

    palavra = input("Digite uma palavra: ").strip().lower()

    # Remover espa√ßos e caracteres especiais (opcional)
    palavra_limpa = ''.join(letra for letra in palavra if letra.isalnum())

    # Inverter a palavra
    palavra_invertida = palavra_limpa[::-1]

    print(f"\nüîç An√°lise da palavra:")
    print(f"   Palavra original: {palavra_limpa}")
    print(f"   Palavra invertida: {palavra_invertida}")

    if palavra_limpa == palavra_invertida:
        print(f"‚úÖ '{palavra}' √â UM PAL√çNDROMO!")
        return True
    else:
        print(f"‚ùå '{palavra}' N√ÉO √â UM PAL√çNDROMO!")
        return False

# Executando o exerc√≠cio 6
e_palindromo = verificar_palindromo()

# ==================== RESUMO DOS RESULTADOS ====================
print("\n" + "=" * 50)
print("üìã RESUMO DA EXECU√á√ÉO")
print("=" * 50)

print("\nüéØ Resultados obtidos em cada exerc√≠cio:")
print(f"1. Concatena√ß√£o realizada com sucesso!")
print(f"2. Texto repetido conforme solicitado!")
print(f"3. Opera√ß√µes matem√°ticas conclu√≠das!")
print(f"4. N√∫mero verificado como {resultado_par_impar}!")
print(f"5. M√©dia calculada: {media_calculada:.2f}")
print(f"6. Pal√≠ndromo: {'SIM' if e_palindromo else 'N√ÉO'}")

print("\n" + "=" * 50)
print("üèÅ TODOS OS EXERC√çCIOS CONCLU√çDOS COM SUCESSO!")
print("=" * 50)
```
